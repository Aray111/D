```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.randn(5, 4), index=[2019, 2020, 2021, 2022, 2023], columns=['W', 'X', 'Y', 'Z'])

print("Original DataFrame:")
print(df)

grouped_df = df.groupby(df.index).sum()

print("\nGrouped DataFrame:")
print(grouped_df)
```

```python
import pandas as pd

data = {'tovar': ['A', 'B', 'A', 'B', 'C'],
        'Value': [10, 15, 20, 25, 30]}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

grouped_stats = df.groupby("tovar").describe().transpose()

print("\nSummary statistics for each 'tovar' group:")
print(grouped_stats)
```

```python
import pandas as pd

data = {'tovar name': ['A', 'B', 'C', 'D', 'E'],
        '2019': [100, 150, 200, 250, 300],
        '2020': [120, 160, 180, 220, 280],
        '2022': [110, 140, 210, 260, 320]}

tips = pd.DataFrame(data)

df1 = tips.iloc[0:20, :][['tovar name', '2019']]
df2 = tips[['tovar name', '2019', '2022']]
df3 = tips.iloc[0:20, :][['tovar name', '2019', '2020']]

print("DataFrame df1:")
print(df1.head())

print("\nDataFrame df2:")
print(df2.head())

print("\nDataFrame df3:")
print(df3.head())
```

```python
df1.plot.area()
```

```python
df1.plot.barh(stacked=True)
```

```python
df1.plot.density()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Date': pd.date_range('2023-01-01', periods=5, freq='D'),
        '2019': [100, 150, 200, 250, 300],
        '2022': [110, 140, 210, 260, 320]}

df2 = pd.DataFrame(data)

df2.set_index('Date', inplace=True)

print("DataFrame df2:")
print(df2)

df2.plot.line(figsize=(12, 3), lw=1, marker='o')
plt.title('Line Plot for df2')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df1.set_index('tovar name').plot.bar(stacked=True, figsize=(8, 6))
plt.title('Stacked Bar Plot for df1')
plt.xlabel('tovar name')
```
